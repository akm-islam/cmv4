# TSNE

```python
import pandas as pd
import plotly.express as px
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

df=pd.read_csv("./data/March.csv")
df_by_topic=df[df['topic_number']==1].reset_index()
corpus = df_by_topic['cleaned_tweet'].to_list()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
df_by_topic["embedding"]=df_by_topic["cleaned_tweet"].apply(lambda x:vectorizer.transform([x]).toarray()[0])
X_embedded = TSNE(n_components=2, learning_rate='auto',init='random').fit_transform(X.toarray())
for i in range(0,len(X_embedded)):
    df_by_topic.at[i, 'tsne_x']=X_embedded[i][0]
    df_by_topic.at[i, 'tsne_y']=X_embedded[i][1]

kmeans = KMeans(n_clusters=10, random_state=0).fit(X.toarray())
df_by_topic["cluster_id"]=df_by_topic["embedding"].apply(lambda x:kmeans.predict([x])[0])

import plotly.express as px
fig = px.scatter(df_by_topic, x="tsne_x", y="tsne_y",color="topic_number")
fig.show()
```

# MDS

```python
import numpy as np
from sklearn.manifold import MDS   
np.random.seed(1867)                               # set the random number seed, so we all get the same answer
n_components = 2                                   # p, reduced dimensionality space
embedding = MDS(n_components=2,n_init = 20,max_iter = 1000,random_state = 73073) # instantiate and set the hyperparameter
MDS_transformed = embedding.fit_transform(X.toarray())

df_by_topic['MDS1'] = MDS_transformed[:,0]
df_by_topic['MDS2'] = MDS_transformed[:,1]

import plotly.express as px
fig = px.scatter(df_by_topic, x="MDS1", y="MDS2",color="topic_number")
fig.show()
```
